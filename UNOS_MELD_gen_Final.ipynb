```python
# process_data.py
import os
import json
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
# Removed import of QuantileTransformer

def process_and_save_data(df, data_dir):
    """
    Process and save data with proper transformations.
    This version applies median imputation for numeric columns and replaces
    NaN values in Encephalopathy_Status and Ascites_Status with "None".

    NOTE: The quantile transformation has been removed to avoid double-normalizing.
    The training pipeline will handle normalization.
    """
    data_dir = Path(data_dir)
    data_dir.mkdir(parents=True, exist_ok=True)

    # Define columns based on your new data structure:
    numerical_columns = [
        "Age",
        "BMI",
        "Na",
        "Cr",
        "INR",
        "Bili",
        "Albumin",
        "MELD",
        "tte"
    ]
    categorical_columns = [
        "Gender",
        "ABO",
        "Dialysis",
        "Etiology",
        "Ethnicity",
        "diab_group_labeled",
        "Encephalopathy_Status",
        "Ascites_Status"
    ]

    # Process missing values:
    # For numeric columns, impute missing values with the median.
    df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
    # For specific categorical columns, replace NaN with "None"
    df["Encephalopathy_Status"] = df["Encephalopathy_Status"].fillna("None")
    df["Ascites_Status"] = df["Ascites_Status"].fillna("None")

    # Split data (using 'outcome' for stratification)
    train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42, stratify=df["outcome"])
    val_df, test_df = train_test_split(temp_df, test_size=0.67, random_state=42, stratify=temp_df["outcome"])

    splits = {
        "train": train_df,
        "val": val_df,
        "test": test_df
    }

    # Save raw numerical and categorical data without applying quantile transformation.
    for split_name, split_df in splits.items():
        np.save(data_dir / f"X_num_{split_name}.npy", split_df[numerical_columns].values)
        np.save(data_dir / f"X_cat_{split_name}.npy", split_df[categorical_columns].values)
        np.save(data_dir / f"y_{split_name}.npy", split_df["outcome"].values)

    # Save info.json with updated feature counts
    info = {
        "name": "meld",
        "id": "meld--default",
        "task_type": "binclass",
        "n_num_features": len(numerical_columns),  # 9
        "n_cat_features": len(categorical_columns),  # 8
        "test_size": len(test_df),
        "train_size": len(train_df),
        "val_size": len(val_df),
        "numerical_columns": numerical_columns,
        "categorical_columns": categorical_columns
    }

    with open(data_dir / "info.json", "w") as f:
        json.dump(info, f, indent=4)

    print("\nDataset split sizes:")
    print(f"Train: {len(train_df)} samples")
    print(f"Val: {len(val_df)} samples")
    print(f"Test: {len(test_df)} samples")
    print("\nNumerical data ranges (raw values):")
    for split_name, split_df in splits.items():
        data = split_df[numerical_columns].values
        print(f"{split_name}: {data.min():.3f} to {data.max():.3f}")

    # No transformer is returned since normalization will be applied during training.
    return None

def verify_and_save_csv(data_dir):
    """
    Verify the processed data files and save them as CSVs.
    """
    data_dir = Path(data_dir)
    with open(data_dir / "info.json", "r") as f:
        info = json.load(f)

    numerical_columns = info["numerical_columns"]
    categorical_columns = info["categorical_columns"]

    print("Verifying data files...")
    splits = ["train", "val", "test"]
    for split in splits:
        print(f"\nChecking {split} split:")
        try:
            X_num = np.load(data_dir / f"X_num_{split}.npy", allow_pickle=True)
            X_cat = np.load(data_dir / f"X_cat_{split}.npy", allow_pickle=True)
            y = np.load(data_dir / f"y_{split}.npy", allow_pickle=True)
            print(f"Numerical shape: {X_num.shape}")
            print(f"Categorical shape: {X_cat.shape}")
            print(f"Target shape: {y.shape}")
            # Create a DataFrame and save as CSV for inspection
            df_num = pd.DataFrame(X_num, columns=numerical_columns)
            df_cat = pd.DataFrame(X_cat, columns=categorical_columns)
            df = pd.concat([df_num, df_cat], axis=1)
            df["outcome"] = y
            csv_path = data_dir / f"{split}_processed.csv"
            df.to_csv(csv_path, index=False)
            print(f"Saved to {csv_path}")
        except Exception as e:
            print(f"Error processing {split} split: {e}")

    print("\nVerification complete.")
    return info

if __name__ == "__main__":
    # Update the path to your CSV file as needed.
    df = pd.read_csv('/content/drive/MyDrive/synthetic_meld_data/UNOS_procesed.csv')
    process_and_save_data(df, "data/meld")
    verify_and_save_csv("data/meld")
```

```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp, chi2_contingency

# --- Load configuration/info ---
# Load the info.json to get the column names
data_dir = "data/meld"  # folder containing your original processed files
with open(os.path.join(data_dir, "info.json"), "r") as f:
    info = json.load(f)

num_cols = info["numerical_columns"]
cat_cols = info["categorical_columns"]

# --- Load Original Dataset ---
# Assuming the original processed CSV was saved (e.g., from process_data.py)
original_csv_path = os.path.join(data_dir, "train_processed.csv")
original_df = pd.read_csv(original_csv_path)

# --- Load Synthetic Dataset ---
# Adjust these paths to wherever your synthetic files were saved (e.g., from sample.py)
synthetic_parent_dir = "exp/meld/check"  # change if needed
synthetic_num_path = os.path.join(synthetic_parent_dir, "X_num_train.npy")
synthetic_cat_path = os.path.join(synthetic_parent_dir, "X_cat_train.npy")

synthetic_num = np.load(synthetic_num_path, allow_pickle=True)
# If categorical data was generated, load it; otherwise, set to None.
if os.path.exists(synthetic_cat_path):
    synthetic_cat = np.load(synthetic_cat_path, allow_pickle=True)
    synthetic_cat_df = pd.DataFrame(synthetic_cat, columns=cat_cols)
else:
    synthetic_cat_df = None

# Create DataFrame for synthetic numerical data
synthetic_num_df = pd.DataFrame(synthetic_num, columns=num_cols)

# Merge synthetic numerical and categorical (if available)
if synthetic_cat_df is not None:
    synthetic_df = pd.concat([synthetic_num_df, synthetic_cat_df], axis=1)
else:
    synthetic_df = synthetic_num_df.copy()

# --- Compare Numerical Variables ---
print("Comparing Numerical Variables:")
for col in num_cols:
    plt.figure(figsize=(10, 4))
    sns.kdeplot(original_df[col], label="Original", fill=True, common_norm=False, alpha=0.5)
    sns.kdeplot(synthetic_df[col], label="Synthetic", fill=True, common_norm=False, alpha=0.5)
    plt.title(f'Distribution Comparison: {col}')
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.legend()
    plt.show()

    # Perform Kolmogorov-Smirnov test
    stat, p_val = ks_2samp(original_df[col].dropna(), synthetic_df[col].dropna())
    print(f"{col}: KS statistic = {stat:.4f}, p-value = {p_val:.4f}\n")

# --- Compare Categorical Variables ---
if synthetic_cat_df is not None:
    print("Comparing Categorical Variables:")
    for col in cat_cols:
        plt.figure(figsize=(10, 4))

        # Calculate normalized frequency counts
        orig_counts = original_df[col].value_counts(normalize=True).sort_index()
        synth_counts = synthetic_df[col].value_counts(normalize=True).sort_index()

        # Combine counts into a DataFrame for plotting
        comp_df = pd.DataFrame({
            'Original': orig_counts,
            'Synthetic': synth_counts
        }).fillna(0)

        comp_df.plot(kind='bar', alpha=0.7)
        plt.title(f'Category Distribution: {col}')
        plt.ylabel("Frequency")
        plt.xlabel(col)
        plt.show()

        # Prepare data for Chi-squared test:
        # For chi-squared, we need the counts (not proportions)
        orig_counts_abs = original_df[col].value_counts().sort_index()
        synth_counts_abs = synthetic_df[col].value_counts().sort_index()

        # Ensure the same index (categories) in both series
        all_categories = sorted(set(orig_counts_abs.index) | set(synth_counts_abs.index))
        orig_counts_abs = orig_counts_abs.reindex(all_categories, fill_value=0)
        synth_counts_abs = synth_counts_abs.reindex(all_categories, fill_value=0)

        contingency_table = pd.DataFrame({
            'Original': orig_counts_abs,
            'Synthetic': synth_counts_abs
        })
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        print(f"{col}: Chi2 statistic = {chi2:.4f}, p-value = {p:.4f}\n")
else:
    print("No synthetic categorical data available to compare.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp, chi2_contingency

def compare_distributions(real_df, syn_df, numerical_cols, categorical_cols):
    # First, align column names
    column_mapping = {
        'INIT_AGENDER': 'GENDER',
        'outcome': 'outcome_encoded'
    }
    syn_df = syn_df.rename(columns=column_mapping)

    # ---------------------------
    # Compare Numerical Variables
    # ---------------------------
    print("=" * 80)
    print("NUMERICAL VARIABLES COMPARISON")
    print("=" * 80)

    for col in numerical_cols:
        if col not in real_df.columns or col not in syn_df.columns:
            print(f"Skipping {col} - not found in both datasets")
            continue

        real_vals = real_df[col].dropna()
        syn_vals = syn_df[col].dropna()
        print(f"\n--- {col} ---")
        print("Real Data Statistics:")
        print(real_vals.describe())
        print("\nSynthetic Data Statistics:")
        print(syn_vals.describe())

        # KS test
        ks_stat, ks_p = ks_2samp(real_vals, syn_vals)
        print(f"KS test: statistic = {ks_stat:.4f}, p-value = {ks_p:.4f}")

        # Plot histograms and KDE
        plt.figure(figsize=(10, 4))
        sns.histplot(real_vals, kde=True, color='blue', label='Real', stat='density', alpha=0.5)
        sns.histplot(syn_vals, kde=True, color='red', label='Synthetic', stat='density', alpha=0.5)
        plt.title(f"Distribution of {col}")
        plt.legend()
        plt.show()

    # ----------------------------
    # Compare Categorical Variables
    # ----------------------------
    print("\n" + "=" * 80)
    print("CATEGORICAL VARIABLES COMPARISON")
    print("=" * 80)

    for col in categorical_cols:
        if col not in real_df.columns or col not in syn_df.columns:
            print(f"Skipping {col} - not found in both datasets")
            continue

        print(f"\n--- {col} ---")

        # Convert to string type to ensure consistent handling
        real_df[col] = real_df[col].astype(str)
        syn_df[col] = syn_df[col].astype(str)

        real_counts = real_df[col].value_counts()
        syn_counts = syn_df[col].value_counts()
        real_props = real_df[col].value_counts(normalize=True)
        syn_props = syn_df[col].value_counts(normalize=True)

        # Create a combined frequency table with all categories
        all_categories = sorted(set(real_counts.index).union(syn_counts.index))
        freq_table = pd.DataFrame({
            'Real': real_counts.reindex(all_categories, fill_value=0),
            'Synthetic': syn_counts.reindex(all_categories, fill_value=0)
        })

        print("\nFrequency Table:")
        print(freq_table)

        # Chi-square test on the frequency table
        try:
            chi2, p_val, dof, expected = chi2_contingency(freq_table.T)
            print(f"\nChi-square test: chi2 = {chi2:.4f}, p-value = {p_val:.4f}")
        except:
            print("\nCould not perform chi-square test")

        # Plot proportions as a bar chart
        plt.figure(figsize=(10, 5))
        props_table = pd.DataFrame({
            'Real': real_props.reindex(all_categories, fill_value=0),
            'Synthetic': syn_props.reindex(all_categories, fill_value=0)
        })

        # Plot side by side bars
        x = np.arange(len(all_categories))
        width = 0.35
        plt.bar(x - width/2, props_table['Real'], width, label='Real', alpha=0.5)
        plt.bar(x + width/2, props_table['Synthetic'], width, label='Synthetic', alpha=0.5)

        plt.xlabel('Categories')
        plt.ylabel('Proportion')
        plt.title(f'Distribution of {col}')
        plt.xticks(x, all_categories, rotation=45, ha='right')
        plt.legend()
        plt.tight_layout()
        plt.show()

# Define columns
numerical_columns = [
    'INIT_AGE', 'INIT_BMI_CALC', 'INIT_SERUM_SODIUM', 'INIT_SERUM_CREAT',
    'INIT_INR', 'INIT_BILIRUBIN', 'INIT_ALBUMIN', 'INIT_MELD', 'time_to_event'
]

categorical_columns = [
    'GENDER', 'ABO', 'Etiology', 'Ethnicity', 'diab_group_labeled',
    'Encephalopathy_Status', 'Ascites_Status', 'INIT_DIALYSIS_PRIOR_WEEK', 'outcome_encoded'
]

# Compare distributions
compare_distributions(real_df, syn_df, numerical_columns, categorical_columns)
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import torch
from torch import nn
from torch.utils.data import DataLoader

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
# Define conditioning columns (order matters)
cond_cols = ['INIT_AGE', 'INIT_BMI_CALC', 'INIT_SERUM_SODIUM', 'INIT_SERUM_CREAT',
             'INIT_INR', 'INIT_BILIRUBIN', 'INIT_ALBUMIN', 'INIT_MELD',
             'INIT_AGENDER', 'ABO', 'Etiology', 'Ethnicity',
             'diab_group_labeled', 'Encephalopathy_Status', 'Ascites_Status',
             'INIT_DIALYSIS_PRIOR_WEEK', 'outcome']

# Separate out conditioning features and target
# Our target column is "time_to_event"
syn_target = syn_df["time_to_event"].values  # leave as is for now
syn_cond = syn_df[cond_cols].copy()

# Decide which columns are numeric and which are categorical.
numeric_features = ['INIT_AGE', 'INIT_BMI_CALC', 'INIT_SERUM_SODIUM',
                      'INIT_SERUM_CREAT', 'INIT_INR', 'INIT_BILIRUBIN',
                      'INIT_ALBUMIN', 'INIT_MELD']
categorical_features = ['INIT_AGENDER', 'ABO', 'Etiology', 'Ethnicity',
                        'diab_group_labeled', 'Encephalopathy_Status',
                        'Ascites_Status', 'INIT_DIALYSIS_PRIOR_WEEK', 'outcome']

# Fit a scaler on the numeric features.
scaler = StandardScaler()
syn_numeric = syn_cond[numeric_features].astype(float)
syn_numeric_scaled = scaler.fit_transform(syn_numeric)

# One-hot encode the categorical features.
# Use drop_first=False to retain all categories.
syn_cat = syn_cond[categorical_features].astype(str)
syn_cat_encoded = pd.get_dummies(syn_cat, drop_first=False)

# Combine numeric and categorical processed features.
syn_features = np.hstack([syn_numeric_scaled, syn_cat_encoded.values])
print("Final synthetic features shape:", syn_features.shape)

# Save the one-hot encoded categorical column names for later manual processing.
cat_feature_names = syn_cat_encoded.columns
```

```python
# Define manual input (all 17 variables except time_to_event)
manual_input = {
    "INIT_AGE": 55,                        # numeric
    "INIT_BMI_CALC": 29.5,                   # numeric
    "INIT_SERUM_SODIUM": 130.0,              # numeric
    "INIT_SERUM_CREAT": 1.2,                 # numeric
    "INIT_INR": 2.5,                       # numeric
    "INIT_BILIRUBIN": 5.0,                   # numeric
    "INIT_ALBUMIN": 3.0,                     # numeric
    "INIT_MELD": 29,                       # numeric
    "INIT_AGENDER": "F",                   # categorical (gender)
    "ABO": "O",                            # categorical
    "Etiology": "Alcohol",                 # categorical
    "Ethnicity": "White",                  # categorical
    "diab_group_labeled": "No Diabetes",   # categorical
    "Encephalopathy_Status": "Grade 3-4",    # categorical
    "Ascites_Status": "Moderate",            # categorical
    "INIT_DIALYSIS_PRIOR_WEEK": "N",       # categorical (assume "N" or "Y")
    "outcome": "0"                         # categorical (set outcome as '0')
}

# Create a one-row DataFrame from manual input.
manual_df = pd.DataFrame(manual_input, index=[0])

# Process numeric part using the same scaler as syn_df.
manual_numeric = manual_df[numeric_features].astype(float)
manual_numeric_scaled = scaler.transform(manual_numeric)

# Process categorical part: one-hot encode then reindex using the same columns as in syn_cat_encoded.
manual_cat = manual_df[categorical_features].astype(str)
manual_cat_encoded = pd.get_dummies(manual_cat, drop_first=False)
manual_cat_encoded = manual_cat_encoded.reindex(columns=cat_feature_names, fill_value=0)

# Combine numeric and categorical parts.
manual_features = np.hstack([manual_numeric_scaled, manual_cat_encoded.values])
print("Manual feature vector shape:", manual_features.shape)
```

```python
# Fit the NearestNeighbors model on synthetic features.
nbrs = NearestNeighbors(n_neighbors=100, algorithm='auto').fit(syn_features)
distances, indices = nbrs.kneighbors(manual_features)

print("Indices of the 100 nearest synthetic patients:")
print(indices[0])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

def compare_distributions(real_df, syn_df, numerical_cols=None, categorical_cols=None):
    """
    Comprehensive comparison of distributions between real and synthetic data
    """
    if numerical_cols is None:
        numerical_cols = real_df.select_dtypes(include=[np.number]).columns

    if categorical_cols is None:
        categorical_cols = real_df.select_dtypes(include=['object', 'category']).columns

    # Set up the plotting style
    plt.style.use('seaborn-v0_8')

    # 1. Statistical Summary for Numerical Variables
    print("="*80)
    print("NUMERICAL VARIABLES COMPARISON")
    print("="*80)

    stats_comparison = pd.DataFrame()
    for col in numerical_cols:
        real_stats = real_df[col].describe()
        syn_stats = syn_df[col].describe()

        # Calculate additional statistics
        ks_stat, p_value = stats.ks_2samp(real_df[col], syn_df[col])
        stats_dict = {
            'Real_Mean': real_stats['mean'],
            'Syn_Mean': syn_stats['mean'],
            'Real_Std': real_stats['std'],
            'Syn_Std': syn_stats['std'],
            'Real_Median': real_stats['50%'],
            'Syn_Median': syn_stats['50%'],
            'KS_Statistic': ks_stat,
            'KS_p_value': p_value
        }
        stats_comparison = pd.concat([stats_comparison,
                                    pd.DataFrame(stats_dict, index=[col])])

    # Print statistical comparison
    pd.set_option('display.float_format', lambda x: '%.3f' % x)
    print("\nStatistical Comparison:")
    print(stats_comparison)

    # 2. Visual Comparison - Numerical Variables
    for col in numerical_cols:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

        # Histogram
        sns.histplot(data=real_df, x=col, alpha=0.5, label='Real', ax=ax1)
        sns.histplot(data=syn_df, x=col, alpha=0.5, label='Synthetic', ax=ax1)
        ax1.set_title(f'Distribution of {col}')
        ax1.legend()

        # Q-Q Plot
        stats.probplot(real_df[col], dist="norm", plot=ax2)
        ax2.set_title(f'Q-Q Plot for {col}')

        plt.tight_layout()
        plt.show()

        # Print range comparison
        print(f"\n{col} Range Comparison:")
        print(f"Real: [{real_df[col].min():.2f}, {real_df[col].max():.2f}]")
        print(f"Synthetic: [{syn_df[col].min():.2f}, {syn_df[col].max():.2f}]")

    # 3. Correlation Matrix Comparison
    print("\n" + "="*80)
    print("CORRELATION COMPARISON")
    print("="*80)

    real_corr = real_df[numerical_cols].corr()
    syn_corr = syn_df[numerical_cols].corr()

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    sns.heatmap(real_corr, annot=True, fmt='.2f', ax=ax1, cmap='coolwarm')
    ax1.set_title('Real Data Correlations')

    sns.heatmap(syn_corr, annot=True, fmt='.2f', ax=ax2, cmap='coolwarm')
    ax2.set_title('Synthetic Data Correlations')

    plt.tight_layout()
    plt.show()

    # 4. Categorical Variables Comparison (if any)
    if len(categorical_cols) > 0:
        print("\n" + "="*80)
        print("CATEGORICAL VARIABLES COMPARISON")
        print("="*80)

        for col in categorical_cols:
            real_prop = real_df[col].value_counts(normalize=True)
            syn_prop = syn_df[col].value_counts(normalize=True)

            print(f"\n{col} Proportions:")
            comparison = pd.DataFrame({
                'Real': real_prop,
                'Synthetic': syn_prop
            }).fillna(0)

            print(comparison)

            # Chi-square test
            chi2, p_value = stats.chi2_contingency(
                pd.crosstab(real_df[col], columns=['Real'])[:1],
                pd.crosstab(syn_df[col], columns=['Synthetic'])[:1]
            )[:2]

            print(f"\nChi-square test: statistic={chi2:.3f}, p-value={p_value:.3f}")

# Usage example:

compare_distributions(
    real_df=df,  # Original dataframe
    syn_df=syn_df,  # Synthetic dataframe
    numerical_cols=['INIT_AGE', 'INIT_BMI_CALC', 'INIT_SERUM_SODIUM', 'INIT_SERUM_CREAT',
                   'INIT_INR', 'INIT_BILIRUBIN', 'INIT_ALBUMIN', 'INIT_MELD', 'time_to_event'],
        categorical_cols=['GENDER', 'ABO', 'Etiology', 'Ethnicity',
       'diab_group_labeled', 'Encephalopathy_Status', 'Ascites_Status',
       'INIT_DIALYSIS_PRIOR_WEEK','outcome_encoded']
)
```

```python
from sklearn.preprocessing import StandardScaler

# Define the feature columns
numeric_cols = ["INIT_AGE", "INIT_BMI_CALC", "INIT_BILIRUBIN",
                "INIT_SERUM_CREAT", "INIT_INR", "INIT_SERUM_SODIUM", "INIT_ALBUMIN"]
categorical_cols = ["GENDER", "ABO", "Etiology", "Ethnicity",
                    "diab_group_labeled", "Encephalopathy_Status", "Ascites_Status", "INIT_DIALYSIS_PRIOR_WEEK"]

# Extract numeric and categorical parts from the synthetic dataframe
syn_numeric = syn_df[numeric_cols]
syn_categorical = syn_df[categorical_cols]

# Fit a scaler on the numeric features
scaler = StandardScaler()
syn_numeric_scaled = scaler.fit_transform(syn_numeric)

# One-hot encode the categorical features (keep all categories; adjust drop_first as desired)
syn_cat_encoded = pd.get_dummies(syn_categorical, drop_first=False)

# Combine the processed features into a final feature DataFrame
import numpy as np
syn_features = np.hstack([syn_numeric_scaled, syn_cat_encoded.values])
print("Final synthetic features shape:", syn_features.shape)

# Save the one-hot column names for later (manual input processing)
cat_feature_names = syn_cat_encoded.columns
```

```python
# Define manual input for each variable (as in your original data)
manual_input = {
    "INIT_AGE": 57,
    "INIT_BMI_CALC": 31.48,
    "INIT_BILIRUBIN": 28.14,
    "INIT_SERUM_CREAT": 2.66,
    "INIT_INR": 2.10,
    "INIT_SERUM_SODIUM": 137.0,
    "INIT_ALBUMIN": 3.30,
    "GENDER": "F",
    "ABO": "O",
    "Etiology": "Alcohol",
    "Ethnicity": "White",
    "diab_group_labeled": "No Diabetes",
    "Encephalopathy_Status": "Grade 1-2",  # or "none" if applicable
    "Ascites_Status": "Slight",
    "INIT_DIALYSIS_PRIOR_WEEK": "N"  # or "Y" depending on your data coding
}

# Create a one-row DataFrame from manual input
manual_df = pd.DataFrame(manual_input, index=[0])

# Process numeric part using the same scaler as for syn_df
manual_numeric = manual_df[numeric_cols].astype(float)
manual_numeric_scaled = scaler.transform(manual_numeric)

# Process categorical part: one-hot encode using pd.get_dummies, then reindex to match syn_cat_encoded columns
manual_categorical = manual_df[categorical_cols].astype(str)
manual_cat_encoded = pd.get_dummies(manual_categorical, drop_first=False)
# Reindex to ensure all columns appear, fill missing with 0
manual_cat_encoded = manual_cat_encoded.reindex(columns=cat_feature_names, fill_value=0)

# Combine numeric and categorical parts
manual_features = np.hstack([manual_numeric_scaled, manual_cat_encoded.values])
print("Manual feature vector shape:", manual_features.shape)
```

```python
from sklearn.neighbors import NearestNeighbors

# Set up NearestNeighbors to find 100 nearest neighbors
nbrs = NearestNeighbors(n_neighbors=100, algorithm='auto').fit(syn_features)
distances, indices = nbrs.kneighbors(manual_features)

print("Nearest neighbor indices (first 10):", indices[0][:10])
```

```python
# Get the 100 nearest synthetic patients from syn_df using the indices
neighbors_df = syn_df.iloc[indices[0]].copy()
neighbors_df.reset_index(drop=True, inplace=True)

print("Nearest 100 synthetic patients:")
neighbors_df.head(10)  # display first 10 rows for brevity
```
